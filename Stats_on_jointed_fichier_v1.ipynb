```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_jointed = pd.read_csv('Jointed_ficher_v1.csv')
```

```python
df_jointed.head()
```

```python
# Convertion des numeros des jours en chaine de caractères
def numToday(x):
    if x==1:
        return "Monday"
    if x==2:
        return "Tuesday"
    if x==3:
        return "Wednesday"
    if x==4:
        return "Thursday"
    if x==5:
        return "Friday"
    if x==6:
        return "Saturday"
    if x==7:
        return "Sunday"
```

```python
# Convertion des numeros des mois en chaine de caractères
def numToMonth(x):
    if x==1:
        return "January"
    if x==2:
        return "February"
    if x==3:
        return "March"
    if x==4:
        return "April"
    if x==5:
        return "May"
    if x==6:
        return "June"
    if x==7:
        return "July"
    if x==8:
        return "August"
    if x==9:
        return "September"
    if x==10:
        return "October"
    if x==11:
        return "November"
    if x==12:
        return "December"
```

```python
df_jointed['Month']= df_jointed['Month'].apply(lambda x : numToMonth(x) )
```

```python
df_jointed
```

```python
#Affichage du nombre d'occurrence de chaque élement de la colonne "UniqueCarrier" dans le dataset principal 2
print(df_jointed['UniqueCarrier'].value_counts().sum())
print(df_jointed['UniqueCarrier'].value_counts())
plt.figure(figsize=(12,8))
sns.countplot(x='UniqueCarrier', data= df_jointed)
```

```python
#Le nombre de minutes de retard par mois et par compagnies
resultat=df_jointed.loc[:, ['UniqueCarrier', 'Month','ArrDelay']].groupby(['Month','UniqueCarrier']).sum()
resultat.columns=['Temps de retard']
resultat=resultat.reset_index()
resultat
```

```python
sns.factorplot(x="UniqueCarrier", y="Temps de retard", hue="Month", data=resultat, kind="bar",height=5, aspect=3)
```

```python
#Nombre de minutes de retard de CarrierDelay par compagnie sur l'année 2008
result=df_jointed.loc[:, ['UniqueCarrier','CarrierDelay']].groupby('UniqueCarrier').sum()
result.columns=['CarrierDelay']
result=result.reset_index()
result
```

```python
sns.catplot(x="UniqueCarrier", y="CarrierDelay", data=result, kind="bar")
```

```python
#Nombre de minutes de retard SecurityDelay par compagnie sur l'année 2008
result1=df_jointed.loc[:, ['Origin','SecurityDelay']].groupby('Origin').sum()
result1.columns=['SecurityDelay']
result1=result1.reset_index()
result1
```

```python
#Nombre de minutes de retard ArrDelay par compagnie sur l'année 2008
result2=df_jointed.loc[:, ['Origin','ArrDelay']].groupby('Origin').sum()
result2.columns=['ArrDelay']
result2=result2.reset_index()
result2
```

```python
plt.figure(figsize=(20,5))
sns.catplot(x="Origin", y="SecurityDelay", data=result1, kind="bar",height=5, aspect=3)
```

```python
plt.figure(figsize=(20,5))
sns.catplot(x="Origin", y="ArrDelay", data=result2, kind="bar",height=5, aspect=3)
```
