```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('2008.csv')
df_slice=pd.read_csv('2008_1.csv')
```

```python
# Taille du dataset principal 1
print("la taille du dataset principal 1:", df.shape[0], "lignes et", df.shape[1],"colonnes")

#Taille du dataset principal 2
print("la taille du dataset principal 2:", df_slice.shape[0], "lignes et", df_slice.shape[1],"colonnes")
```

```python
# Convertion des numeros des jours en chaine de caractères
def numToday(x):
    if x==1:
        return "Monday"
    if x==2:
        return "Tuesday"
    if x==3:
        return "Wednesday"
    if x==4:
        return "Thursday"
    if x==5:
        return "Friday"
    if x==6:
        return "Saturday"
    if x==7:
        return "Sunday"
```

```python
# Convertion des numeros des mois en chaine de caractères
def numToMonth(x):
    if x==1:
        return "January"
    if x==2:
        return "February"
    if x==3:
        return "March"
    if x==4:
        return "April"
    if x==5:
        return "May"
    if x==6:
        return "June"
    if x==7:
        return "July"
    if x==8:
        return "August"
    if x==9:
        return "September"
    if x==10:
        return "October"
    if x==11:
        return "November"
    if x==12:
        return "December"
```

```python
df_slice['DayOfWeek']= df_slice['DayOfWeek'].apply(lambda x : numToday(x) )
df_slice['Month']= df_slice['Month'].apply(lambda x : numToMonth(x) )
df_slice['FlightNum']=df_slice['FlightNum'].astype(object)
df_slice['DepTime']=df_slice['DepTime'].astype(object)
df_slice['CRSDepTime']=df_slice['CRSDepTime'].astype(object)
df_slice['ArrTime']=df_slice['ArrTime'].astype(object)
df_slice['CRSArrTime']=df_slice['CRSArrTime'].astype(object)
```

```python
#Description de chaque colonne du dataset principal 2

for col in df_slice.columns:
    print("\nDescription de la colonne",col)
    print(df_slice[col].describe())
```

```python
# Créer une nouvelle fonction qui détermine si la valeur en paramètre est manquante:
def num_missing(x):
    return sum(x.isnull())
```

```python
# On applique cette fonction au dataset principal 2 pour chaque colonne:
print("Nombre de valeurs manquantes par colonne:")
print(df_slice.apply(num_missing, axis=0)) #axis=0 définit que la fonction sera bien appliquée sur chaque colonne
```

```python
#Le nombre de minutes de retard par mois et par compagnies

resultat=df_slice.loc[:, ['UniqueCarrier', 'Month','ArrDelay']].groupby(['Month','UniqueCarrier']).sum()
resultat.columns=['Temps de retard']
resultat
```

```python
#Nombre de minutes de retard par compagnie sur l'année 2008
result=df_slice.loc[:, ['UniqueCarrier','ArrDelay']].groupby('UniqueCarrier').sum()
result.columns=['Tenmps de retard']
result
```

```python
#Classement des compagnies en fonction du temps moyen de retard sur l'année 2008

temp1=df_slice.loc[:, ['UniqueCarrier','ArrDelay']].groupby('UniqueCarrier').sum()
temp1.columns=['Temps de retard']
temp2=pd.DataFrame(df_slice['UniqueCarrier'].value_counts())
temp2.columns=['Nombre de vols']
concatenation=temp1.join(temp2)
concatenation['Temps moyen de retard']=concatenation['Temps de retard']/concatenation['Nombre de vols']
concatenation['Rang'] = concatenation['Temps moyen de retard'].rank()

concatenation=concatenation.reset_index() #transforme un index en colonne et réindexe en partant de 0 à n-1
sns.factorplot(x='UniqueCarrier',y='Temps moyen de retard', data= concatenation, kind='bar',size=10)
concatenation
```
